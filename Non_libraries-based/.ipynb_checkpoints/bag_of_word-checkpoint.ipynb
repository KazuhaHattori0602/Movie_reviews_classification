```python
#Make vocabulary
def bag_of_words(path):
    from nltk.stem.wordnet import WordNetLemmatizer
    lmtzr = WordNetLemmatizer()
    list_files = CD.all_files(path)
    list_of_text = [] #Tap hop cac van ban
    vocab_set = set({})
    print ('Processing bag of word...')
    file_stopwords = open(r'..\stopwords.txt', 'r+')
    stopwords = file_stopwords.read()
    count = 1
    for file in list_files:
        print (count)
        count += 1
        f = open(file, 'r+')
        data = f.read()
        small_BoW = data.split()
        small_BoW = [lmtzr.lemmatize(word) for word in small_BoW]
        #Remove words with <= 2 chars and stopwords and Lemmatizer
        for word in small_BoW:
            if word in stopwords or len(word) <= 2:
                small_BoW.remove(word)
        list_of_text.append(small_BoW)
        vocab_set = vocab_set.union(set(small_BoW))
        f.close()
    file_stopwords.close()
    vocab_list = [word for word in vocab_set]
    list_of_words_id = []
    count = 1
    for text in list_of_text:
        print (count)
        count += 1
        words_id = []
        for word in text:
            words_id.append(vocab_list.index(word))
        list_of_words_id.append(words_id)
    print ('Done bag of word')
    return vocab_list, list_of_words_id
```
